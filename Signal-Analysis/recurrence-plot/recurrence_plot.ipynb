```python
def slenkantis_listas(a, window, step_size):
    shape = a.shape[:-1] + (a.shape[-1] - window + 1 - step_size + 1, window)
    strides = a.strides + (a.strides[-1] * step_size,)
    arr = np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)
    row_mask = (arr != 0.0).all(axis=1)
   #print(arr[row_mask,:])
    return arr[row_mask,:]
```

```python
def euklidine_norma(skirtumas):
    kvadratu = [i ** 2 for i in skirtumas]
    return math.sqrt(sum(kvadratu)) 
    
def manheteno_norma(skirtumas):
    return(sum(abs(skirtumas)))

def maksimumo_norma(skirtumas):
    return(max(abs(skirtumas)))
```
